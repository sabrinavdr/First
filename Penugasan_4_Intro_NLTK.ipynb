```python
#Case Folding

kalimat = "Pemuda berumur 27 tahun. Tidur dengan sepatu masih di kaki. Baru pulang shubuh tadi. Terlalu lelah. Terlalu penat. Terlalu sesak. Terlalu..... Begitu saja hidupnya tiga tahun terakhir. Macam kalong. Tidur di siang hari. Berjaga penuh di malam hari. Menghabiskan dingin dan lengangnya malam sambil menggerutu di kedai minuman, bar tengah kota. Duduk di pojok ruangan. Sendirian. Menatap galak ke siapa saja yang mencoba basabasi bertegur sapa (termasuk waitress genit yang mengantarkan botol bir). Gesture wajah dan gerakan tubuhnya jelas sekali: Pergi!! Biarkan aku sendiri! Matahari semakin meninggi. Pemuda kita masih tertidur."
lower_case = kalimat.lower()
print(lower_case)
```

```python
#menghapus tanda baca
import string
kalimat2 = lower_case.translate(str.maketrans("","",string.punctuation))
print(kalimat2)
```

```python
#penghapusan kata stopword
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

import string
token = word_tokenize(kalimat2)
listStopword = set(stopwords.words('indonesian'))

removed = []
for t in token:
  if t not in listStopword:
    removed.append(t)
    print(removed)
```

```python
#mencari frekuensi kemunculan token
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import string
kalimat3 = "'pemuda', 'berumur', '27', 'tidur', 'sepatu', 'kaki', 'pulang', 'shubuh', 'lelah', 'penat', 'sesak', 'hidupnya', 'kalong', 'tidur', 'siang', 'berjaga', 'penuh', 'malam', 'menghabiskan', 'dingin', 'lengangnya', 'malam', 'menggerutu', 'kedai', 'minuman', 'bar', 'kota', 'duduk', 'pojok', 'ruangan', 'menatap', 'galak', 'mencoba', 'basabasi', 'bertegur', 'sapa', 'waitress', 'genit', 'mengantarkan', 'botol', 'bir', 'gesture', 'wajah', 'gerakan', 'tubuhnya', 'pergi', 'biarkan', 'matahari', 'meninggi', 'pemuda', 'tertidur'"
kalimat4 = kalimat3.translate(str.maketrans("","",string.punctuation))
tokens = nltk.tokenize.word_tokenize(kalimat4)
kemunculan = nltk.FreqDist(tokens)
print(kalimat4)
print(tokens)
print(kemunculan.most_common())

print(kalimat4)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed = []
for token in tokens:
  stemmed_word = stemmer.stem(token)
  stemmed.append(stemmed_word)

print(stemmed)
```
