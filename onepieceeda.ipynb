### Filter % symbol from Manga% and Anime%

```python
def filter_percent(data):
    return data[:-1]
```

```python
data['Manga%'] = data['Manga%'].apply(filter_percent).astype(float)
data['Anime%'] = data['Anime%'].apply(filter_percent).astype(float)
```
